#### Performing Calculations on ints and floats

##### Addition

```python
num1 = 3
num2 = 4.5

result = num1 + num2

print(result)

#shorthand version
result += 2.5 #result = result + 2.5

print(result)
```

##### Subtraction

```python
result_diff = num2 - num1

print(result_diff)

#shorthand
result_diff -= 0.5
print(result_diff)
```

##### Multiplication

```python
result_mul = num1 * num2

print(result_mul)

#shorthand
result_mul *= 2
print(result_mul)
```

##### Division

```python
result_div = num2 / num1

print(result_div)

#shorthand
result_div /= 2
print(result_div)
```

##### Floor Division

```python
#drops the numbers after decimal & gives whole number
result_floor = num2 // num1

print(result_floor)

#shorthand
result_floor //= 2
print(result_floor)
```

##### Modulo

```python
#divides and then gives a remainder only
result_mod = num2 % 1 
print(result_mod)

```

##### Exponential

```python
square = 5 ** 2
print(square)

#shorthand
square **= 2
print(square)
```

### If Statements

```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num1 = 5
num2 = 10

if num2 == num1:
    print('Equal Values')
else:
    print('Not Equal')
    
if num2 >= num1:
    print('Num2 is greater')
else:
    print('Num1 is greater')
```

### While Loops

```python
#if you dont set a condition or break, while loop will run forever!
#while, condition statment
#continues based on conditional statment

num = 0

while num < 10:
    print(num)
    num += 1
```

##### Looping 'While True'

```python
#loops over conditions and breaks with a break or when condition is True
#useful when working with user input and you want to continue receiving inputs(until they hit stop condition)

game_over = True

while True:
    print('infinite loop')
    if game_over == False:
        break
```

##### While & For Loops Used Together

```python
num = 5

while num < 5:
    print('\n While Loop Iteration: ' + str(num))
    for i in range(2):
        print('For Loop Iteration: ', str(i))
    num += 1
```

### Lists

##### Declaring Lists

```python
l_1 = [] #empty lst declaration

names = ['Bob', 'Sarah', 'Timmy', 'Gabby', 'Tina']
print(names)
```

##### Indexing a List

```python
#to index use [] with a start and stop and step
# [start : stop : step]
#to move backwards use a negative [-1]

#single index
print(names[0])

#print starting at index 1 going to the end
print(names[1:])

#print starting at the beginning and going to position n -- (n is number we want to stop at)
print(names[:2])

#print starting at index 1 going by 2 to the end
print(names[1::2])

#print starting from the end in reverse
print(names[::-1])
```

##### .append()

```python
#adds to the end
names.append('Laura')
print(names)
```

##### .insert()

```python
#list.insert(index,value)

names.insert(3, 'Jon')
print(names)

print(names.pop(3))
```

##### .pop()

```python
#default to the last value if no parameter is given
#pop returns the element that was removed--this is good if you want to assign it to a variable

my_name = names.pop()
print(my_name)
(print(names))

```


##### .remove()

##### Concatenating Two Lists

```python
# this will append two lists together, not add up their elements

l_1 = [0,1,2]
l_2 = [3,4,5]

large_list = l_1 + l_2

print(large_list)
```

##### Lists Within Lists

```python
#lists are able to hold any element, including other LISTS
#they can go as deep as you want, having many lists inside of eachother

names = ['Max', 'Sam', 'Josh', ['Sally', 'Sue', 'Tameka']]
print(names)
print(names[3][1])
```

##### Looping Through Lists

```python
#two ways to loop through lists one is by index, the other is by using the 'in' keyword

#loop with a for loop by index
for i in range(len(names)):
    print(names[i])
    
#loop with for keyword using 'in'
for i in names:
    print(i)
```

```python
#looping using enumerate
#gives the index and value at the same time
for index,value in enumerate(names):
    print('the index is: ', index, 'the value is: ', value)
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didn’t
# An else after a for runs if the for didn’t break

for n in range(1,101):
    if n > 1:
        for j in range(2,n):
            if(n % j == 0):
                break
        else:
            print(n)
    
    
```
